```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def predator_prey(t, z, alpha, beta, delta, gamma):
    D, R = z
    dDdt = alpha * D - beta * D * R
    dRdt = delta * D * R - gamma * R
    return [dDdt, dRdt]
```

```python
D0 = 500
R0 = 500
```

```python
alpha = 0.05   # Growth rate of Democrats
beta = 0.0001  # Influence of Republicans on Democrats
delta = 0.0001 # Influence of Democrats on Republicans
gamma = 0.04   # Decline rate of Republicans
```

```python
t_span = (0, 50)
t_eval = np.linspace(0, 50, 100)
```

```python
solution = solve_ivp(predator_prey, t_span, [D0, R0], t_eval=t_eval, args=(alpha, beta, delta, gamma))
```

```python
plt.figure(figsize=(8,5))
plt.plot(solution.t, solution.y[0], label="Democrat Votes")
plt.plot(solution.t, solution.y[1], label="Republican Votes")
plt.xlabel("Time")
plt.ylabel("Votes")
plt.legend()
plt.title("Predator-Prey Model for Competing Parties")
plt.show()
```
