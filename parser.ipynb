```python
import pandas as pd
```

```python
df = pd.read_csv("SENATE_precinct_general.csv")
```

```python
df.head
```

```python
exclude_values = ["OVERVOTES", "WRITEIN", "UNDERVOTES", 'TIMES BLANK VOTED']
exclude_mode = ['ABSENTEE', 'EARLY', 'PROVISIONAL', 'VOTE CENTER',
       'EARLY TSX', 'EARLY TSX FLE', 'FLE TSX', 'FLE VC', 'FLW TSX',
       'FLW VC', 'POLLING', 'TC TSX', 'TC VC1', 'TC VC2', 'TSX']
filtered_df = df[~df["candidate"].isin(exclude_values)]
filtered_df = filtered_df.loc[filtered_df['state'] == 'ARIZONA']
filtered_df = filtered_df[~filtered_df['mode'].isin(exclude_mode)]
```

```python
filtered_df
```

```python
df = filtered_df.drop(columns=['special', 'writein', 'state_po', 'state_fips', 'state_cen', 'state_ic', 'date', 'readme_check', 'state', 'district', 'magnitude', 'year', 'stage', 'dataverse', 'jurisdiction_fips', 'jurisdiction_name', 'party_simplified', 'office', 'mode'])
df
```

```python
df.to_csv("output.csv", index=False)
```
