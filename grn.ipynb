```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_edges_from([
    ("DEMOCRAT", "REPUBLICAN", {"weight": -0.5}),  # Negative influence
    ("REPUBLICAN", "DEMOCRAT", {"weight": -0.3}),  # Negative influence
    ("GREEN", "DEMOCRAT", {"weight": 0.4}),  # Positive influence
    ("INDEPENDENT", "REPUBLICAN", {"weight": 0.2})  # Positive influence
])
```

```python
plt.figure(figsize=(6,6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', edge_color='gray', font_size=10)
labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Party Influence Network (Gene-Regulatory Model)")
plt.show()
```
